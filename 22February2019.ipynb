```python
def findmax(my_list):
    max=my_list[0]
    s=0
    for i in range(len(my_list)):
        if(max<my_list[i]):
            max=my_list[i]
            s=i
    return max,s
list_1=[1,2,3,4,5,56,71,77,566,234,377,5442,2678]
findmax(list_1)
```

```python
def recursive_fib(n):
    if(n<2):
        return n
    else:
        return recursive_fib(n-1)+recursive_fib(n-2)
```

```python
import time
start=time.time()
for i in range(35):
    
    s=recursive_fib(i)
    print(s)
end=time.time()
print(end-start)
```

```python
def power_1(m,n):
    t=1
    global sayac
    for i in range(n):
        sayac=sayac+1
        t=t*m
    return t,sayac
```

```python
def call_report(x,y):
    global sayac
    sayac=0
    r=power_1(x,y)
    print(x," üzeri",y," degeri",r[0],"çağrım sayısı : ",r[1])
```

```python
def power_2(x,n):
    global sayac
    sayac=sayac+1
    if(n==0):
        return 1,sayac
    elif(n==1):
        return x,sayac
    elif(n%2==0):
        return (power_2(x*x,n//2)[0],power_2(x*x,n//2)[1])
    elif(n%2==1):
        return (power_2(x*x,n//2)[0]*x,power_2(x*x,n//2)[1])
```

```python
def call_report_recursive(x,y):
    global sayac
    sayac=0
    r=power_2(x,y)
    print(x," üzeri",y," degeri",r[0],"çağrım sayısı : ",r[1])
```

```python
call_report_recursive(2,6)
```
